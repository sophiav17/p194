# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
#read the csv
df = pd.read_csv('framingham.csv')
df
```

```python
#Filter and make a new dataframe for non smokers
non_smokers = df.loc[df['currentSmoker'] == 0]
non_smokers
```

```python
#Group by age column and find average heart rate at different age
group_non_smokers = groupby('age')['heartRate'].mean()
group_non_smokers
```

```python
#Filter and make a new dataframe for smokers
smokers = df.loc[df['currentSmoker'] == 1]
smokers
```

```python
#Group by age column and find average heart rate at different age
group_smokers - groupby('age')['heartRate'].mean()
group_smokers
```

```python
import pandas as pd
import matplotlib.pyplot as plt

group_non_smokers = groupby('age')['heartRate'].mean()
group_non_smokers

group_smokers - groupby('age')['heartRate'].mean()
group_smokers

#Plot a line graph to show the heart rate of smokers vs non smokers
plt.figure(figsize = (10,8))
age = group_non_smokers('age')
heart_rate = group_non_smokers('heartRate')

plt.plot(age, heart_rate, label = "Heart Rate of Non Smokers", linewidth = 3.0)

age = group_smokers('age')
heart_rate = group_smokers('heartRate')

plt.plot(age, heart_rate, label = "Heart Rate of Smokers", linewidth = 3.0)

plt.xlabel('Age')
plt.ylabel('Heart Rate')

plt.legend()

plt.title()
plt.show()
```

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('framingham.csv')
df
#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years
coronary_heart_disease = df.loc[df['TenYearCHD'] == 1]
coronary_heart_disease
```

```python
#Group by age column and count the rows of TenYearCHD column
group_age = groupby('age')('TenYearCHD').count().reset_index()
group_age
```

```python
#Plot a line graph to show total number of people having a chance of coronary heart disease in 10 years
plt.figure(figsize = (10,8))
plt.scatter(xaxis = 'Age', yaxis = 'Ten year CHD', label = 'Age group chances of having coronary heart disease in the next 10 years', color = 'teal', alpha = 0.8, s = 3)

plt.xlabel('Age')
plt.ylabel('Ten Year Coronary Heat Disease')

plt.legend()
plt.title()

plt.show()
```
